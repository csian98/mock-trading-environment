```python
import os, sys
sys.path.append("pylib/")
from stock_market import StockMarket
import numpy as np
import math
```

```python
symbols = [
            "NVDA",
            "AAPL",
            "GOOG",
            "MSFT",
            "AMZN",
            "META",
            "TSLA",
            "AVGO",
            "ORCL",
            "NFLX",
        ]
market = StockMarket(symbols)
history = market.history(period='2y', days=60)
```

```python
A = set([])

for symbol in symbols:
    prices = history[symbol]
    returns  =  prices[1:]/prices[:-1] - 1
    returns = returns[~np.isnan(returns)]
    returns = np.abs(returns)
    A.update(returns.tolist())

p = 0.95
c = np.quantile(list(A), p)
print(c)
```
